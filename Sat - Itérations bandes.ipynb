```python
# L2A 10m - B2 et B3 pour calcul BI

import os

folder = "E:\\Images Sat Kerguelen\\2020\\Sentinel-2"

```

```python
###### RECHERCHE D'UNE BANDE UNIQUE ######
# Listes pour bandes
list_T42FWM_band = []
list_T42FWK_band = []
list_T42FXL_band = []
list_T42FWL_band = []
list_T42FVK_band = []
list_T42FVL_band = []
list_T42FVM_band = []
list_autre_band = []
```

```python
# Pour recherche d'une bande unique. Ici B02
for path, subdirs, files in os.walk(folder):
    for name in files:
        if name.endswith("jp2") and "B02" in name and "10m" in name and "2020" in name:
            if "T42FWM" in name: 
                full_name=os.path.join(path,name)
                list_T42FWM_band.append(full_name)
            elif "T42FWK" in name: 
                full_name=os.path.join(path,name)
                list_T42FWK_band.append(full_name)
            elif "T42FXL" in name: 
                full_name=os.path.join(path,name)
                list_T42FXL_band.append(full_name)
            elif "T42FWL" in name:
                full_name=os.path.join(path,name)
                list_T42FWL_band.append(full_name)
            elif "T42FVK" in name: 
                full_name=os.path.join(path,name)
                list_T42FVK_band.append(full_name)
            elif "T42FVL" in name: 
                full_name=os.path.join(path,name)
                list_T42FVL_band.append(full_name)
            elif "T42FVM" in name: 
                full_name=os.path.join(path,name)
                list_T42FVM_band.append(full_name)
            else :
                list_autre_band.append(name)
```
